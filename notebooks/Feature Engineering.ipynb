```python
import pandas as pd

cat_a = pd.read_csv('../data/cat_a.csv')
cat_b = pd.read_csv('../data/cat_b.csv')

df = pd.concat([cat_a, cat_b])

df.head()
```

```python
df.columns
```

```python
df.describe()
```

```python
cols = ['bidding_exercise', 'year']

df.drop(cols, axis=1, inplace=True)

df.head()
```

```python
df = df.sort_values('date', ascending=True)

df.head()
```

```python
df.info()
```

```python
#Feature Engineering on original data

#Convert date field from string to a datetime datatype
df['date'] = pd.to_datetime(df['date'])

#Extract months from date
df['month'] = pd.DatetimeIndex(df['date']).month

df['bid_quota_ratio'] = df['total_bids'] / df['quota']

df.head()
```

```python
df.info()
```

```python
df.to_csv('../data/cat_ab_original_features.csv', index=False)
```

```python
cat_a = pd.read_csv('../data/cat_a_population_cpi.csv')
cat_b = pd.read_csv('../data/cat_b_population_cpi.csv')

df = pd.concat([cat_a, cat_b])

df.head()
```

```python
cols = ['bidding_exercise', 'year']

df.drop(cols, axis=1, inplace=True)

df.head()
```

```python
#Feature engineering on vehicle population data

#Convert date field from string to a datetime datatype
df['date'] = pd.to_datetime(df['date'])

#Extract months from date
df['month'] = pd.DatetimeIndex(df['date']).month

df['bid_quota_ratio'] = df['total_bids'] / df['quota']

#Percentage changes with regards to previous months
df['quota_pct_change'] = df['quota'].pct_change()
df['price_pct_change'] = df['price'].pct_change()
df['total_bids_pct_change'] = df['total_bids'].pct_change()
df['success_bids_pct_change'] = df['success_bids'].pct_change()

#TODO: Calculate Prevailing Quota Premium (PQP) - 3 month prevailing average of COE prices
```

```python
df.dropna(inplace=True)

df.head()
```
